```python
## Function 1

import re
from collections import defaultdict

species_pbdb = defaultdict(list)

infile = "canids/formatted_pbdb_data.csv"
with open(infile, 'r') as ff:
    canid_recs = ff.readlines()[1:]
# 5 is the genus, 8 is age
    
    for line in canid_recs:
        species = line.split('","')[5] #'""' gets rid of quotes
        genus = line.split(",")[6] 
        
        minage = line.split('","')[10] #'""' gets rid of quotes
        maxage = line.split('","')[11]  
        mean_age = (float(minage)+float(maxage))/ 2  #using the mean age as the age
        if re.search(r"\s", species) and not re.search(r"genus", genus):  #this filters for only binomial (using the space between words as identifier) and non-genus species!
            species_pbdb[species] = [mean_age]
print("{" + "\n".join("{}: {}".format(k, v) for k, v in species_pbdb.items()) + "}") #prints dict on new lines
```
